**Problem statement**: to find a relation between the given sets of input and output arrays.

**First approach:**

**proposed solution**: every i th position in the output array is the product of all the elements of the input array except the i th position value of input array.

Here the **time complexity is high** since we multiply 3 numbers in every loop to get one particular element of the output list 

Product_excluding() function is to find the resultant value of every position in the output list(j)

```python
def product_excluding(p,q):
    r_val=1
    for i in p:
        if i!=q:
            r_val=r_val*i
    return r_val
```

output() function is to iterate through the input list and to generate the output list

```python
def output(x):
    j=[]
    for i in x:
        j.append(product_excluding(x,i))
    return j
```

```python
data = [int(item) for item in input("Enter the input array: ").split()]
print("Output:",output(data))
```

**Second approach:**

**proposed solution**: in this method we multiply only once to get the product of all the elements at first, then we just divide this value by the particular element to get the corresponding element in the output array.

The **time complexity is relatively lower** when compared to the first approach.

```python
data = [int(item) for item in input("Enter the input array: ").split()]
product=1
output=[]

for i in data:     #To find the product of all elements 
  product=product*i

for j in data:     #To divide the product by that particular element to get output array
  output.append(int(product/j)) #Here i have used to terminate the decimals that arise due to division operator, just to make it look similar to the output in the assignment sheet

print(output)
```
